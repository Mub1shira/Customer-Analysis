</div>

<div style="color:red;">1. Introduction </div>

 
Customer segmentation is a crucial aspect of modern marketing and business strategy. It involves dividing a customer base into distinct groups with shared characteristics. This documentary report outlines the fundamental aspects of a customer segmentation project, including what customer analysis is, where to obtain data, the project's benefits and uses, and two critical types of segmentation.

<div style="color:red;">2. Customer Analysis </div>

Customer analysis is the process of collecting and analyzing data about customers to gain insights into their behaviors, preferences, and needs. This information is then used to create customer segments, which are groups of customers who exhibit similar traits.

<div style="color:red;">3.Data Sources </div>

To conduct customer segmentation effectively, you need reliable and comprehensive data sources. Here's where to obtain data for your project:





<div style="color:indigo ;">*Internal CRM (Customer Relationship Management) System: "CompanyCRM.com"  

*E-commerce Website Data:  "EcomDataHub.com" 

*Customer Surveys: "SurveyMonkey.com" or "Google Forms" 

*Email Marketing Platforms: "MailChimp.com"  

*Customer Support and Helpdesk Data: "SupportDeskPro.com"  

*Purchase History from Online Marketplaces: Amazon or eBay 

*Website Analytics Tools:"WebTrends.com" </div>



<div style="color:red;">4.Benefits and Use of Customer Segmentation</div> 
Customer segmentation offers several benefits and applications:

<div style="color:indigo;">a.Targeted Marketing: </div>

By understanding customer segments, you can create tailored marketing campaigns that resonate with specific groups, leading to higher engagement and conversion rates.
<div style="color:indigo;">b.Product Development: </div> 

Segmentation can guide product development by identifying which features or offerings are most important to different customer groups.
<div style="color:indigo;">c.Personalization: </div> 

Personalized experiences, both online and offline, can be delivered to customers based on their segment, enhancing customer satisfaction.
<div style="color:indigo;">d.Cost Efficiency: </div> 

Resource allocation becomes more efficient when you know which segments are most profitable. This leads to cost savings and better ROI.

<div style="color:red;">5. Types of Segmentation
Two important types of customer segmentation are:</div>

<div style="color:indigo;">a. Demographic Segmentation: </div>

This divides customers based on demographic factors such as age, gender, income, education, and occupation. For example, a cosmetics company may target young females aged 18-30.
<div style="color:indigo;">b. Behavioral Segmentation: </div>

Behavioral segmentation categorizes customers based on their actions, such as purchase history, website visits, and brand loyalty. For instance, an e-commerce platform might segment customers as 'frequent shoppers' or 'cart abandoners'.

First of all, for the missing values, First check the outlier in the income column

```python
income_column = df['Income']

# Calculate the IQR (Interquartile Range)
Q1 = income_column.quantile(0.25)
Q3 = income_column.quantile(0.75)
IQR = Q3 - Q1

# Define a lower bound and upper bound for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(income_column < lower_bound) | (income_column > upper_bound)]

# Display only the 'Income' column of the outliers
outlier_incomes = outliers['Income']

# Display the outlier incomes
print("Outliers in the Income column:")
print(outlier_incomes)
```

We've got some outliers here. The next step is to remove the extraneous ones

```python
# Remove outliers by filtering the DataFrame
df_no_outliers = df[(income_column >= lower_bound) & (income_column <= upper_bound)]

# Display the DataFrame without outliers
print("DataFrame without outliers:")
print(df_no_outliers['Income'])
```

After removing the outliers, the null value is replaced by the mean value.

```python

# Calculate the mean of the 'Income' column
income_mean = df['Income'].mean()

# Fill null values in the 'Income' column with the calculated mean
df['Income'].fillna(income_mean, inplace=True)

# Now, the null values in the 'Income' column have been replaced with the mean value
```

In the next step, I am going to create a feature out of "Dt_Customer" that indicates the number of days a customer is registered in the firm's database. However, in order to keep it simple, I am taking this value relative to the most recent customer in the record.

Thus to get the values I must check the newest and oldest recorded dates.

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
dates = []
for i in df["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
#Dates of the newest and oldest recorded customer
print("The newest customer's enrolment date in therecords:",max(dates))
print("The oldest customer's enrolment date in the records:",min(dates))
```

```python
#Created a feature "Customer_For"
days = []
d1 = max(dates)  # Assuming 'dates' contains valid date values
for i in dates:
    delta = d1 - i
    days.append(delta)
df["Customer_For"] = days
df["Customer_For"] = pd.to_numeric(df["Customer_For"], errors="coerce")
```

    <h1>Dimensionality Reduction</h1>
</div>



In this problem, there are many factors on the basis of which the final classification will be done. These factors are basically attributes or features. The higher the number of features, the harder it is to work with it. Many of these features are correlated, and hence redundant. This is why I will be performing dimensionality reduction on the selected features before putting them through a classifier.
Dimensionality reduction is the process of reducing the number of random variables under consideration, by obtaining a set of principal variables.

Principal component analysis (PCA) is a technique for reducing the dimensionality of such datasets, increasing interpretability but at the same time minimizing information loss.

Steps in this section:

Dimensionality reduction with PCA
Plotting the reduced dataframe
Dimensionality reduction with PCA

For this project, I will be reducing the dimensions to 3.

```python
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#A 3D Projection Of Data In The Reduced Dimension
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

<div style="color: orange; background-color:white; text-align: left;">
    <h1>Clustering</h1>
</div>

With attributes now reduced to three dimensions, the next phase involves the application of Agglomerative clustering, a hierarchical clustering technique. This method entails the iterative merging of data points until the intended number of clusters is reached.

The clustering process can be broken down into the following key steps:

<div style="color:red;">Utilizing the Elbow Method for Cluster Count Selection:</div>
To establish the appropriate number of clusters, we employ the Elbow Method, a technique that guides us in making an informed choice.

<div style="color:red;">Execution of Clustering Using Agglomerative Clustering:</div>
We put Agglomerative clustering into action, which systematically combines data points in a hierarchical manner, ultimately leading to the desired number of clusters.

<div style="color:red;">Evaluation of Formed Clusters via Scatter Plot Analysis:</div> After the clustering process, we delve into an in-depth examination of the resultant clusters by visualizing them through scatter plots. This visualization aids in comprehending the cluster compositions and their interrelations.






```python
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```


The previous section suggests that the ideal cluster count for this dataset is four. Moving forward, we will proceed with applying the Agglomerative Clustering Model to establish the definitive clusters.

 To calculates the silhouette score for a given clustering model and data

```python
score = silhouette_score(scaled_ds,Elbow_M.labels_, metric='euclidean')
score
```

The silhouette score ranges from -1 to 1, where a higher score indicates that the objects in one cluster are more similar to each other and dissimilar to objects in other clusters.

```python
#Initiating the Agglomerative Clustering model 
AC = AgglomerativeClustering(n_clusters=4)

# fit model and predict clusters
yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = yhat_AC

#Adding the Clusters feature to the orignal dataframe.
df["Clusters"]= yhat_AC
```

```python

data_AC = AC.fit_predict(scaled_dt)                    
scaled_dt["Clusters"] = data_AC
data["Clusters"]= data_AC        
```

```python
score = silhouette_score(scaled_ds,Elbow_M.labels_, metric='euclidean')
score
```

To examine the clusters formed let's have a look at the 3-D distribution of the clusters.

```python
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Define colors for each cluster (adjust the colors as needed)
colors = ['red', 'blue', 'green', 'purple']

# Iterate through each cluster and plot the data points
for cluster_num in range(4):  # we have 4 clusters
    cluster_data = PCA_ds[PCA_ds["Clusters"] == cluster_num]
    ax.scatter(cluster_data['col1'], cluster_data['col2'], cluster_data['col3'],
               c=colors[cluster_num], label=f'Cluster {cluster_num}')

# Label the axes
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]

# Add a legend
ax.legend()

# Show the 3D plot
plt.show()
```
